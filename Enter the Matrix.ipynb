```python
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
from matplotlib import markers as markers
```

```python
df = pd.read_csv('fantasy_summary.csv')
```

```python
df['Rebounds'] = df['OREB'] + df['DREB']
```

```python
plt.figure(figsize=(9,6))
markers = itertools.cycle((',', '+', 'o', '*')) 
colors = np.linspace(0, len(df['Player']))
mymap = plt.get_cmap("tab10")
my_colors = mymap(colors)
for index, row in df.iterrows():
    plt.scatter(df[df['Player'] == row['Player']]['FGM'], df[df['Player'] == row['Player']]['FG%'], label = row['Player'], color = my_colors[index], marker = next(markers), s = 60)
plt.title('FGM to FG%', fontsize = 14)
plt.xlabel('FGM', fontsize = 14)
plt.ylabel('FG%', fontsize = 14)
plt.legend(loc = (1.01, 0.715))
```
