```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('chris_fantasy.csv')
df2 = pd.read_csv('arian_fantasy.csv')
df3 = pd.read_csv('james_fantasy.csv')
df4 = pd.read_csv('harish_fantasy.csv')
df5 = pd.read_csv('daniel_fantasy.csv')
df6 = pd.read_csv('joey_fantasy.csv')
```

```python
# FGM

fig = plt.figure(figsize=(17,10))
ax = fig.add_subplot(111)
ax.plot(df['Week'], df['FGM'].cumsum(), marker = 'o', label = 'Chris', alpha = 0.5)
ax.plot(df2['Week'], df2['FGM'].cumsum(), marker = 'o', label = 'Arian', alpha = 0.5)
ax.plot(df3['Week'], df3['FGM'].cumsum(), marker = 'o', label = 'James', alpha = 0.5)
ax.plot(df4['Week'], df4['FGM'].cumsum(), marker = 'o', label = 'Harish', alpha = 0.5)
ax.plot(df5['Week'], df5['FGM'].cumsum(), marker = 'o', label = 'Daniel', alpha = 0.5)
ax.plot(df6['Week'], df6['FGM'].cumsum(), marker = 'o', label = 'Joey', alpha = 0.5)
plt.yticks(np.arange(0, 4000, 200))
plt.legend()
plt.title('FGM')
```
