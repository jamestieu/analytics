```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('chris_fantasy.csv')
df2 = pd.read_csv('arian_fantasy.csv')
df3 = pd.read_csv('james_fantasy.csv')
df4 = pd.read_csv('harish_fantasy.csv')
df5 = pd.read_csv('daniel_fantasy.csv')
df6 = pd.read_csv('joey_fantasy.csv')
df7 = pd.read_csv('fantasy_weekly.csv')
```

```python
fig = plt.figure(figsize=(17,10))
ax = fig.add_subplot(111)
ax.plot(df['Week'], df['FGM'], marker = 'o', label = 'Chris', alpha = 0.5)
ax.plot(df2['Week'], df2['FGM'], marker = 'o', label = 'Arian', alpha = 0.5)
ax.plot(df3['Week'], df3['FGM'], marker = 'o', label = 'James', alpha = 0.5)
ax.plot(df4['Week'], df4['FGM'], marker = 'o', label = 'Harish', alpha = 0.5)
ax.plot(df5['Week'], df5['FGM'], marker = 'o', label = 'Daniel', alpha = 0.5)
ax.plot(df6['Week'], df6['FGM'], marker = 'o', label = 'Joey', alpha = 0.5)
plt.title('FGM')
# plt.yticks(np.arange(df['FGM'].min() - 10, df['FG%'].max() + 10, step= 100))
plt.xticks(np.arange(df['Week'].min(), df['Week'].max() + 1, 1))
pinny_middle = df7['FGM'].mean()
plt.axhline(y=pinny_middle, label='Average stilllll', linestyle='--', color='orange', alpha = 0.5)
std = df7['FGM'].std()
pinny_top = pinny_middle + std
plt.axhline(y=pinny_top, label='Exceptional!', linestyle='--', color='green', alpha = 0.5)
pinny_bottom = pinny_middle - std
plt.axhline(y=pinny_bottom, label='Terrible!', linestyle='--', color='red', alpha = 0.5)
plt.legend(bbox_to_anchor=(1, 1), loc='best')
```
