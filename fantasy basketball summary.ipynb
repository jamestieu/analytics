```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fantasy_summary.csv')
```

```python
df.head()
```

```python
df.rename(columns= {'Unnamed: 0' : 'Player'}, inplace = True)
```

```python
df.head()
```

```python
df.set_index('Player')
```

```python
# FGM sorted
result = df.sort_values('FGM', ascending = False)
result.head()
```

```python
colors = {'Chris':'blue', 'Arian':'red', 'Harish':'purple', 'James':'orange', 'Daniel':'brown', 'Joey':'green'}
plt.figure(figsize=(10,10))
plt.yticks(np.arange(0, 4000, step=100))
plt.bar(result['Player'], result['FGM'], color = [colors[t] for t in result['Player']])
plt.title('FGM')
```
