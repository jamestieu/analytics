```python
df = pd.read_csv('fantasy_summary.csv')
```

```python
df.rename(columns= {'Unnamed: 0' : 'Player'}, inplace = True)
```

```python
df.set_index('Player')
```

```python
# FGM sorted
result = df.sort_values('FGM', ascending = False)
```

```python
colors = {'Chris':'blue', 'Arian':'red', 'Harish':'purple', 'James':'orange', 'Daniel':'brown', 'Joey':'green'}
plt.figure(figsize=(10,10))
plt.yticks(np.arange(0, 4000, step=100))
plt.bar(result['Player'], result['FGM'], color = [colors[t] for t in result['Player']])
pinny_top = np.percentile(result['FGM'], 75)
plt.axhline(y=pinny_top, label='Tradeable Surplus', linestyle='--', color='blue')
pinny_median = result['FGM'].median()
plt.axhline(y=pinny_median, label='Maybe', linestyle='--', color='orange')
pinny_66 = np.percentile(result['FGM'], 66)
plt.axhline(y=pinny_66, label = 'Good Chance', linestyle='--', color='green')
pinny_33 = np.percentile(result['FGM'], 33)
plt.axhline(y=pinny_33, label = 'Improve or Punt', linestyle='--', color='red')
plt.title('FGM')
plt.legend()
```
